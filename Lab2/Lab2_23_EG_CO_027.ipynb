```python
# Import Project Dependencies
import os
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
import pathlib as path
```

```python
main_dir = path.Path("../Datasets/Dataset")

data_frames = []

for dirpath, dirnames, filenames in os.walk(main_dir):

    root_dir = path.Path(dirpath)

    label = root_dir.name

    for file in filenames:

        if file.endswith(".csv"):

            file_path = root_dir / file

            # print(f"Loading: {file_path}")

            data = pd.read_csv(file_path, low_memory=False)

            data["label"] = label

            data_frames.append(data)

if data_frames:

    data = pd.concat(data_frames, ignore_index=True)

    print("Successfully combined all CSV files.")

    

else:

    print("No CSV files found.")
```

```python
# Drop totally empty columns
# Drop totally empty columns
columns_to_drop = [col for col in data.columns if data[col].isna().all()]
data.drop(columns=columns_to_drop, inplace=True, errors='ignore')
data.drop(columns=columns_to_drop, inplace=True, errors='ignore')
```

```python
# Fix inconsistent datatypes
string_numeric_columns = ["V_A", "V_B", "V_C", "I_A", "I_B", "I_C", "ActivePower", "ReactivePower", "ip.len"]
for col in string_numeric_columns:
    if col in data.columns:
        data[col] = data[col].astype(str).str.replace(",", "")
        data[col] = pd.to_numeric(data[col], errors="coerce")


# Verify final datatypes
data.info()
```

```python
# Handle Duplicate Entries
duplicates = data.duplicated().sum()
print(f"Removing {duplicates} duplicate rows... ")
data.drop_duplicates(inplace=True)
```

```python
# Handle Missing Values safely
categorical_data = ["eth.src", "eth.dst", "ip.srt", "ip.dst", "_ws.col.protocol", "frame.time.epoch"]
for col in categorical_data:
    if col in data.columns:
        data[col] = data[col].fillna("UNKNOWN")
```

```python
# Impute numerical features with median
numerical_col = data.select_dtypes(include=[np.number]).columns
col_to_input = [col for col in numerical_col if col  not in ["node", "PLCID"]]
for col in col_to_input:
    if col in data.columns:
        median_val = data[col].median()
        data[col] = data[col].fillna(median_val)
```

```python
# Drop rows that are complementary to the cyber-physical data
data.dropna(subset=["node", "PLCID"], how="all", inplace=True)
print(f"Datasets shape after drop: {data.shape}")
```

```python
# Isolate predicting feature by dropping network packet metadata
cols_to_exclude = ["frame.time_epoch", "@timestamp", "label", "eth.src",
                   "eth.dst", "ip.src", "ip.dst", "_ws.col.protocol"]
X_features = data.drop(columns=[col for col in cols_to_exclude if col in data.columns])
y_target = data["label"]
print(f"Selected predictor Features ({X_features.shape[1]})")
print(f"Columns : \n{list(X_features.columns)}")
```

```python
# Encode categorical columns
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y_target)
class_mapping = dict(zip(range(len(label_encoder.classes_)), label_encoder.classes_))
print(f"Target classes Recorded: {class_mapping}")
```

```python
# Split into training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_test, y_train = train_test_split(X_features, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)
```

```python
# Generate the descriptive statistics for the training features.
X_train.describe().T
```

```python
# Standardize the training and testing feature matrices using StandardScaler

from sklearn.preprocessing import StandardScaler

# Initialize scaler
scaler = StandardScaler()

# Select only numeric columns
numeric_cols = X_train.select_dtypes(include=[np.number]).columns

# Fit scaler on training data and transform both train and test sets
X_train_scaled_vals = scaler.fit_transform(X_train[numeric_cols])
X_test_scaled_vals = scaler.transform(X_test[numeric_cols])

# Create copies of original DataFrames
X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()

# Replace numeric columns with scaled values
X_train_scaled[numeric_cols] = X_train_scaled_vals
X_test_scaled[numeric_cols] = X_test_scaled_vals

# Verify shapes
print("X_train_scaled shape:", X_train_scaled.shape)
print("X_test_scaled shape:", X_test_scaled.shape)
```
